```python
import pandas as pd 
import numpy as np
import re 
from datetime import datetime
```

```python
df = pd.read_csv('/Users/almaliezenga/Documents/projects/bubble_chart_v4-master/data/tabula-Geschenken 08-11-2023.csv')
```

```python
df = df.T.reset_index().T
df = df.reset_index()
```

```python
ids = []
for i, r in df.iterrows(): 
    name = re.search(".*, .* \(.*\) - .*", r[0])
    if name is not None:
        full_name = name[0]
        print(full_name)
        ids.append(i)
    df.at[i, 'receiver'] = full_name
    
```

```python
df = df.drop(ids)
df = df.reset_index()
```

```python
for i, r in df.iterrows():
    try: 
        date = re.search(".{2}-.{2}-.{4}", r[1])
        print(date)
    except: 
        date = 'nan'
        
    if date is None or date == 'nan' :
        print
        df = df.drop(i)
    elif date is not None: 
        df.at[i, 'date'] = datetime.strptime(date[0], '%d-%m-%Y')
```

```python
df[['receiver', 'party']] = df['receiver'].str.split(' - ', expand=True)
```

```python
df = df[[0, 'receiver', 'date', 'party']]
df = df.reset_index()
df
```

```python
#Opties 
#Ontvangen () van 
#Bijgewoond () op uitnodiging van 
#Deelgenomen aan () () 
forbidden_last_words = ['een', 'twee', '5', 'voor', 'de', 'gelegenheid', 'het', '2', 'tijdens',]

for i, r in df.iterrows(): 
    received = re.search("Ontvangen .*", r[0])
    if received is None:
        df = df.drop(i)
    else: 
        gifter = re.search("van .{5,50}?( een | de | het | twee | voor |ter gelegenheid| één | tijdens | 1 | 5 | 2 | 3 | 4)", r[0])
        if gifter: 
            print(gifter[0])
            lis = list(gifter[0].split(" "))
            last_word = lis[len(lis)-2]
            print(last_word) 
            
            if last_word in forbidden_last_words: 
                df.at[i, 'gifter'] = gifter[0].rsplit(' ', 1)[0].split(' ', 1)[1].rsplit(' ', 1)[0]
            else: 
                df.at[i, 'gifter'] = gifter[0].rsplit(' ', 1)[0].split(' ', 1)[1]
```

```python
for i, r in df.iterrows(): 
    price = re.search("(€|$).{1,5}(,.{1,2}|-})", r[0])
    if price is None: 
        df.at[i, 'price'] = np.nan
    else: 
        df.at[i, 'price'] = price[0].replace('€', '').replace(',', '.').replace('.-.', '').replace('.--', '').replace('.-', '').replace(' e', '')
print(df['price'].isna().sum())
#convert whole column to float
df['price'] = df['price'].apply(pd.to_numeric, errors='coerce')
print(df['price'].isna().sum())
```

```python
for i, r in df.iterrows(): 
    gift = re.search("( een | de | het | twee | voor |ter gelegenheid| één | 5 | 2 | 3 | 4).* (De waarde is onbekend|ter waarde van)", r[0])
    if gift is None: 
        df.at[i, 'gift'] = np.nan
    else: 
        df.at[i, 'gift'] = gift[0]
```

```python
df = df.rename(columns={0: "description"})
```

```python
for i, r in df.iterrows():
    
    try:
        book = re.search("(boek|atlas)", r['description'])
        if book is not None: 
            df.at[i, 'category'] = "boeken"
 
        food = re.search("(etenswaren|Stamppotdiner|maaltijd|fles|bonbons|chocolade|koek|couscous|Vlaaien|visschotel|kersen|knoopjesdrop|kinderlunchbox|ministroopwafels)", r['description'])
        if food is not None: 
            df.at[i, 'category'] = "etenswaren en dranken"

        ticket = re.search("(toegangskaart|toegangsbewijs|bezoek|kaart|uitnodiging|ticket)", r['description'])
        if ticket is not None: 
            df.at[i, 'category'] = "uitnodigingen en tickets"
    
        flowers = re.search("bloemen", r['description'])
        if flowers is not None: 
            df.at[i, 'category'] = "bloemen"
    except: 
        df.at[i, 'category'] = np.nan
```

```python
df.to_csv('/Users/almaliezenga/Documents/projects/bubble_chart_v4-master/data/cleaned_geschenkregister.csv')
```
